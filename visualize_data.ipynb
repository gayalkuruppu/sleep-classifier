```python
import pandas as pd

hr_p46343 = pd.read_csv('outputs/features/46343_hr_feature.out')
count_p46343 = pd.read_csv('outputs/features/46343_count_feature.out')
cosine_p46343 = pd.read_csv('outputs/features/46343_cosine_feature.out')
time_p46343 = pd.read_csv('outputs/features/46343_time_feature.out')
psg_label_p46343 = pd.read_csv('outputs/features/46343_psg_labels.out')
```

```python
import matplotlib.pyplot as plt

# visualize hr feature against time feature
plt.plot(time_p46343, hr_p46343)
plt.xlabel('time (hours)')
plt.ylabel('heart rate (smoothened)')
plt.title('Heart rate of a sample patient')
plt.show()
```

```python
# visualize count feature against time feature

plt.plot(time_p46343, count_p46343)
plt.xlabel('time (hours)')
plt.ylabel('step count (movement)')
plt.title('Count of a sample patient')
plt.show()
```

```python
# visualize cosine feature against time feature
plt.plot(time_p46343, cosine_p46343)
plt.xlabel('time (hours)')
plt.ylabel('cosine')
plt.title('Circadian Rythm approximation of a sample patient')
plt.show()
```

```python
# visualize psg labels against time feature
plt.plot(time_p46343, psg_label_p46343)
plt.xlabel('time (hours)')
plt.ylabel('psg label')
plt.title('PSG labels of a sample patient')
plt.show()
```
